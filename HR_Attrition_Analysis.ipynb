```python
# INSIGHT - Correlation Heatmap:
#
# Attrition_Num correlations (last row):
# - No numeric variable has strong correlation with Attrition.
# - Highest negative correlations with Attrition are:
#   JobLevel (-0.17), MonthlyIncome (-0.16), TotalWorkingYears (-0.17)
#   meaning higher income and seniority = less attrition.
# - Age (-0.16) and YearsAtCompany (-0.13) also negatively correlated.
# - This confirms chi-square findings — seniority and income
#   are the strongest numeric predictors of attrition.
#
# Multicollinearity findings:
# - JobLevel and MonthlyIncome are highly correlated (0.95)
#   — they measure almost the same thing.
# - TotalWorkingYears is strongly correlated with JobLevel (0.78),
#   MonthlyIncome (0.77) and YearsAtCompany (0.63).
# - YearsAtCompany, YearsInCurrentRole, YearsWithCurrManager
#   are all highly correlated with each other (above 0.70).
# - PercentSalaryHike and PerformanceRating are correlated (0.77).
#
# CONCLUSION:
# - Attrition is driven more by categorical factors than numeric ones.
# - If building a model, drop MonthlyIncome or JobLevel (redundant).
# - Drop YearsInCurrentRole or YearsWithCurrManager (redundant).
```

```python
from scipy.stats import chi2_contingency #chi2 ise used for categorical categorical calculations
# list of categorical variables to test
columns = ['JobRole', 'Department', 'OverTime', 'MaritalStatus',
           'Gender', 'Education', 'BusinessTravel', 'AgeGroup',
           'IncomeGroup', 'YearsGroup', 'JobSatisfaction',
           'WorkLifeBalance', 'EnvironmentSatisfaction',
           'JobLevel', 'NumCompaniesWorked', 'PerformanceRating']

print(f"{'Variable':<25} {'Chi2':>10} {'P_Value':>12} {'Significant':>12}")
print("-" * 62)

for col in columns:
    try:
        ct=pd.crosstab(df[col],df['Attrition'])
        chi2,p_value,dof,expected=chi2_contingency(ct)
        significant='Yes' if p_value<0.05 else 'No'
        print(f"{col:<25} {chi2:>10.2f} {p_value:>12.4f} {significant:>12}")
    except:
        print(f"{col:<25} {'Error':>10}")
```

### INSIGHT - Chi-Square Test Results:
#
# SIGNIFICANT variables (p < 0.05) - 13 out of 16:
# - JobRole, OverTime, JobLevel, IncomeGroup have highest
#   chi2 scores meaning they are strongest attrition drivers.
# - AgeGroup, YearsGroup, MaritalStatus, BusinessTravel
#   are also strongly significant.
# - JobSatisfaction, WorkLifeBalance, EnvironmentSatisfaction,
#   Department, NumCompaniesWorked are moderately significant.
#
# NOT SIGNIFICANT variables (p > 0.05) - 3 out of 16:
# - Gender (p=0.2906) - no meaningful impact on attrition.
# - Education (p=0.5455) - education level does not drive attrition.
# - PerformanceRating (p=0.9901) - completely irrelevant variable.
#
# CONCLUSION:
# - 13 variables are statistically proven to influence attrition.
# - Gender, Education and PerformanceRating can be dropped
#   from further analysis.
